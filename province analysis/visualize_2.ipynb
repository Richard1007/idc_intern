## 3.计算区县的省内GDP占比

```python
df_merge["省内GDP占比"] = np.nan
df_merge["省内人口占比"] = np.nan
```

```python
prov_gdp_dict = {}
prov_population_dict = {}
for province in province_lst:
  prov_gdp_dict[province] = df_merge.loc[df_merge['Province of City (省)'] == province, 'GDP'].sum()
  prov_population_dict[province] = df_merge.loc[df_merge['Province of City (省)'] == province, '人口（万）'].sum()
```

```python
# df_merge = df_merge.reset_index()  # make sure indexes pair with number of rows
GDP_lst = []
population_lst = []
for index, row in df_merge.iterrows():
  province = row['Province of City (省)']
  population = row['人口（万）']
  GDP = row['GDP']
  # print("{0:.1%}".format(GDP / prov_gdp_dict[province]))
  # GDP_lst.append("{0:.1%}".format(GDP / prov_gdp_dict[province]))
  # population_lst.append("{0:.1%}".format(population / prov_population_dict[province]))
  GDP_lst.append(GDP / prov_gdp_dict[province])
  population_lst.append(population / prov_population_dict[province])
```

```python
df_merge["省内GDP占比"] = GDP_lst
df_merge["省内人口占比"] = population_lst
```

```python
plt.rcParams.update({'font.size': 6}) # must set in top
'''
Args: bar_1,bar_2,line_1,line_2,district_lst are simple lists
'''
def combo_chart(province,order_by,bar_1,bar_2,line_1,line_2,district_lst):  
  bar_name_1 = '城镇人均收入'
  bar_name_2 = '农村人均收入'
  line_name_1 = '人口省内占比'
  line_name_2 = 'GDP省内占比'

# Create a data df
  df_plot = pd.DataFrame({
  bar_name_1 : bar_1,
  bar_name_2 : bar_2,
  line_name_1 : line_1,
  line_name_2 : line_2
   })
  # print(df_plot)
  # df = pd.melt(df, id_vars="class", var_name="sex", value_name="survival rate")
  # line_columns = df_plot.columns[2:]

  # Bar Chart
  width = .65 # width of a bar
  df_plot[[bar_name_2,bar_name_1]].plot(kind='bar', width = width,ylim=(0,80000))

  # df_plot.plot(style={line_name_1: 'r', line_name_2: 'g'},secondary_y=True)
  # df_plot[line_columns].plot.line(secondary_y=True,y='hello')
  # Line Chart
  df_plot[line_name_1].plot(kind='line',secondary_y=True,color = 'r')
  df_plot[line_name_2].plot(kind='line',secondary_y=True,color = 'y')

  # ax2.set_ylabel('Y2 data', color='b')
  # Show percentage in y-axis
  ax = plt.gca()
  ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))

  
  ax.set_xlabel('x-axis')
  # Annottaion
  # plt.text(.91, .89, 'Red: 人口省内占比 Yellow：GDP省内占比', ha='left', va='top', transform=ax.transAxes)

  # X labels
  ax.set_xticklabels((i for i in district_lst))
  # plt.ylabel('黄色：GDP省内占比；红色：人口省内占比', fontsize=8, rotation=-90)
  title = province+' '+ order_by+' '+'top 10'
  plt.title(title)
  # plt.show()
  plt.savefig('11.13/'+title+'.png')
```
